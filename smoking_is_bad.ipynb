# Is smoking really good?
Now it is up to you to run the analysis further and check what could be underlying your cousings results.

```python
#Create new column with 0 and 1 instead of "no" and "yes" for Smoker Column
mapping = {'Yes': 1, 'No': 0}
df_clean['smoker_int'] = df_clean['smoker'].map(mapping)

#Create a new column with 0 and 1 instead of "female" and "male" for the Gender Column
mapping_2 = {"male":1, "female": 0}
df_clean['gender_int'] = df_clean['gender'].map(mapping_2)

#Remove Null Rows 
df_clean = df_clean.dropna()
df_clean
#Choose the Columns the run the regression
x = df_clean[["smoker_int", "age","salary", "gender_int"]]
y = df_clean[["alive"]]

#Run the Regression 
model = LinearRegression()
model.fit(x, y)

print(model.coef_.round(6))
```

When controlling for age, salary and gender, we can see that the impact of smoking on the probability of being alive is now much smaller at 0.017 compared to approximately 0.08 before. It is still however interesting that the coefficient is still positive. 

```python
df_restricted = df_clean[df_clean["age"] <=55]

x = df_restricted[["smoker_int", "age","salary", "gender_int"]]
y = df_restricted[["alive"]]

#Run the Regression 
model = LinearRegression()
model.fit(x, y)

print(model.coef_.round(5))
```

Now, if we run the same regression, but this time only look at individuals who are younger than 55 years old, we can see the negative relationship between smoking and your probability of being alive. This indicates that the table likely includes a lot of individuals who were old and hence died, but did not smoke, which biased the results.
